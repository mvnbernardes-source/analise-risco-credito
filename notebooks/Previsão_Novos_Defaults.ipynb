```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
novos_clientes = pd.read_csv('base_prevista.csv')
```

```python
novos_clientes.head()
```

```python
# padronizando os dados
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
novos_clientes_scaled = scaler.fit_transform(novos_clientes)
```

```python
# Dando download no modelo treinado
import joblib
modelo = joblib.load('random_forest_model.pkl')
```

```python
# prevendo os novos defaults
previsoes = modelo.predict(novos_clientes_scaled)
novos_clientes['previsao_default'] = previsoes
```
